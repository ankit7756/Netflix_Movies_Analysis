## Cleaning Plan

### Columns to drop entirely:
- duration: 16,000 nulls, completely empty, no analytical value
- show_id: unique identifier only, no analytical value
- description: text field, not needed for quantitative analysis
- type: dataset is all movies, column has no variation

### Zeros to replace with NaN (disguised missing data):
- budget: 11,153 zeros = 69.7% missing financial data
- revenue: 10,355 zeros = 64.7% missing financial data
  → These will be replaced with NaN so they don't corrupt averages
  → All financial analysis will use a separate financial_df subset

### Zeros to KEEP (not missing data):
- vote_count: 894 zeros = genuinely unrated movies
- vote_average: 899 zeros = genuinely unrated movies  
- rating: 899 zeros = same movies
  → These are real values meaning "no rating exists"
  → Will filter these out only when doing rating analysis

### Real nulls to fill:
- director: 132 nulls → fill with 'Unknown'
- cast: 204 nulls → fill with 'Unknown'  
- country: 466 nulls → fill with 'Unknown'

### Rows to drop:
- genres: 107 null rows → drop entirely
  (only 0.67% of data, genre is essential for analysis)

### Data type fixes:
- date_added: object → datetime
- language: strip whitespace and lowercase for consistency

```python
# Drop useless columns
df.drop(columns=['show_id', 'duration', 'description', 'type'], inplace=True)
print("Columns remaining:", df.columns.tolist())
```

```python
# Replace financial zeros with NaN:
import numpy as np

df['budget'] = df['budget'].replace(0, np.nan)
df['revenue'] = df['revenue'].replace(0, np.nan)

# Verify
print("Budget nulls now:", df['budget'].isnull().sum())
print("Revenue nulls now:", df['revenue'].isnull().sum())
```

```python
# Fix date type:
df['date_added'] = pd.to_datetime(df['date_added'])

# Verify
print(df['date_added'].dtype)
print(df['date_added'].head())
```

```python
#  Standardize text columns:
df['language'] = df['language'].str.strip().str.lower()
df['country'] = df['country'].str.strip()

# Verify no weird spaces remain
print(df['language'].value_counts().head(5))
```

```python
# Fill real nulls:
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# Verify
print("Director nulls:", df['director'].isnull().sum())
print("Cast nulls:", df['cast'].isnull().sum())
print("Country nulls:", df['country'].isnull().sum())
```

```python
# Drop genre null rows:
rows_before = len(df)
df = df.dropna(subset=['genres'])
rows_after = len(df)

print(f"Rows dropped: {rows_before - rows_after}")
print(f"Rows remaining: {rows_after}")
```

```python
# Final verification:
print("=== AFTER CLEANING ===")
print(df.isnull().sum())
print(f"\nFinal shape: {df.shape}")
```

## Phase 2 — Feature Engineering

Raw data rarely contains everything needed for deep analysis. 
Feature engineering creates new meaningful columns derived from 
existing ones — turning raw numbers into analytical tools.

In this phase we create:
- Financial metrics: profit, ROI, flop flag
- Budget categorization: grouping movies by production scale
- Rating quality tiers: translating scores into categories
- Time features: extracting year, month from dates
- Genre explosion: splitting multi-genre strings for genre-level analysis

All financial features are created on a separate subset (financial_df) 
containing only movies with verified budget AND revenue data.

```python
# Create verified financial subset — only movies with real budget and revenue
financial_df = df[df['budget'].notna() & df['revenue'].notna()].copy()

print(f"Full dataset: {len(df):,} movies")
print(f"Financial subset: {len(financial_df):,} movies")
print(f"Coverage: {len(financial_df)/len(df)*100:.1f}% of total library")
```

```python
# Calculate profit — absolute money made after costs
financial_df['profit'] = financial_df['revenue'] - financial_df['budget']

# Calculate ROI percentage — return relative to investment
financial_df['roi_pct'] = (
    (financial_df['revenue'] - financial_df['budget']) 
    / financial_df['budget'] * 100
)

# Flag flops — movies that failed to recoup their budget
financial_df['was_flop'] = financial_df['revenue'] < financial_df['budget']

# Verify
print(financial_df[['title', 'budget', 'revenue', 'profit', 'roi_pct', 'was_flop']].head())
print(f"\nFlop rate: {financial_df['was_flop'].mean()*100:.1f}%")
```

```python
# Categorize movies by production scale
financial_df['budget_tier'] = pd.cut(
    financial_df['budget'],
    bins=[0, 10_000_000, 50_000_000, 100_000_000, float('inf')],
    labels=['Low (<$10M)', 'Mid ($10-50M)', 'High ($50-100M)', 'Blockbuster ($100M+)']
)

# Check distribution
print(financial_df['budget_tier'].value_counts())
print(financial_df['budget_tier'].value_counts(normalize=True).mul(100).round(1))
```

```python
# Categorize vote average into quality tiers
df['rating_tier'] = pd.cut(
    df['vote_average'],
    bins=[0, 5, 6.5, 8, 10],
    labels=['Poor', 'Average', 'Good', 'Excellent']
)

# Check distribution — exclude unrated (vote_average == 0)
rated = df[df['vote_average'] > 0]
print(rated['rating_tier'].value_counts())
print(rated['rating_tier'].value_counts(normalize=True).mul(100).round(1))
```

```python
# Extract time components from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
df['movie_age'] = 2025 - df['release_year']

# Verify
print(df[['title', 'date_added', 'year_added', 'month_added', 'movie_age']].head())
```

```python
# Split multi-genre strings into individual rows for genre-level analysis
df_genres = df.assign(
    genre=df['genres'].str.split(', ')
).explode('genre')

# Check how many unique genres exist
print(f"Unique genres: {df_genres['genre'].nunique()}")
print("\nTop 15 genres by movie count:")
print(df_genres['genre'].value_counts().head(15))
```

```python
# Verify all new columns created correctly
print("=== MAIN DATAFRAME NEW COLUMNS ===")
print(df[['rating_tier', 'year_added', 'month_added', 'movie_age']].head())
print(f"\nMain df shape: {df.shape}")

print("\n=== FINANCIAL DATAFRAME NEW COLUMNS ===")
print(financial_df[['profit', 'roi_pct', 'was_flop', 'budget_tier']].head())
print(f"\nFinancial df shape: {financial_df.shape}")

print("\n=== GENRE DATAFRAME ===")
print(f"Genre df shape: {df_genres.shape}")
print(f"Unique genres: {df_genres['genre'].nunique()}")
```

## Phase 3 — EDA and Business Questions

With clean data and engineered features in place, we now answer 
8 specific business questions a real investor or content strategist 
would care about. Each question is answered with code, numbers, 
and a written interpretation.

The goal is not just to produce outputs but to extract meaning — 
every finding is followed by what it actually suggests for 
decision making.

Three dataframes in use:
- df — general library analysis
- financial_df — all financial questions
- df_genres — all genre-level analysis

```python
# Question 1 — What does the Netflix library look like overall?

# Release year distribution — last 15 years
print("=== MOVIES BY RELEASE YEAR (Last 15 years) ===")
print(df['release_year'].value_counts().sort_index().tail(15))

# How many movies added to Netflix per year
print("\n=== MOVIES ADDED TO NETFLIX PER YEAR ===")
print(df.groupby('year_added')['title'].count())

# Language breakdown
print("\n=== TOP 10 LANGUAGES ===")
print(df['language'].value_counts().head(10))
```

```python
# Question 2 — Which genres dominate the library vs which make the most money?

# Genre count in full library
print("=== TOP 10 GENRES BY VOLUME ===")
print(df_genres['genre'].value_counts().head(10))

# Genre average ROI — financial subset only
print("\n=== AVERAGE ROI BY GENRE ===")
financial_genres = financial_df.assign(
    genre=financial_df['genres'].str.split(', ')
).explode('genre')

print(
    financial_genres.groupby('genre')['roi_pct']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .round(1)
)
```

```python
# Question 3 — Does bigger budget mean bigger revenue?

# Correlation between budget, revenue and rating
print("=== CORRELATION MATRIX ===")
print(financial_df[['budget', 'revenue', 'vote_average']].corr().round(3))

# Average revenue by budget tier
print("\n=== AVERAGE REVENUE BY BUDGET TIER ===")
print(financial_df.groupby('budget_tier')['revenue'].mean().apply(lambda x: f"${x:,.0f}"))

# Average ROI by budget tier
print("\n=== AVERAGE ROI BY BUDGET TIER ===")
print(financial_df.groupby('budget_tier')['roi_pct'].mean().round(1))
```

```python
# Question 4 — What does a flop look like vs a success?

# Overall flop rate
print("=== OVERALL FLOP RATE ===")
print(financial_df['was_flop'].value_counts(normalize=True).mul(100).round(1))

# Flop rate by budget tier
print("\n=== FLOP RATE BY BUDGET TIER ===")
print(financial_df.groupby('budget_tier')['was_flop'].mean().mul(100).round(1))

# Average audience rating — flops vs successes
print("\n=== AVERAGE RATING: FLOPS VS SUCCESSES ===")
print(financial_df.groupby('was_flop')['vote_average'].mean().round(2))
```

```python
# Question 5 — Which languages produce commercially successful non-English movies?

# Filter non-English with real financial data
non_english = financial_df[financial_df['language'] != 'en']

# Average revenue by language
print("=== AVERAGE REVENUE BY LANGUAGE (Non-English) ===")
print(
    non_english.groupby('language')['revenue']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .apply(lambda x: f"${x:,.0f}")
)

# Movie count with verified financial data by language
print("\n=== MOVIE COUNT WITH FINANCIAL DATA BY LANGUAGE ===")
print(
    non_english.groupby('language')['title']
    .count()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
# Question 6 — Has movie quality changed over the years?

# Average rating by release year — last 15 years
print("=== AVERAGE RATING BY RELEASE YEAR ===")
rated = df[df['vote_average'] > 0]
print(rated.groupby('release_year')['vote_average'].mean().tail(15).round(2))

# Average popularity by release year
print("\n=== AVERAGE POPULARITY BY RELEASE YEAR ===")
print(df.groupby('release_year')['popularity'].mean().tail(15).round(2))
```

```python
# Question 7 — Which directors produce the highest rated content?

# Directors with 3+ movies sorted by average rating
print("=== TOP DIRECTORS BY AVERAGE RATING (min 3 movies) ===")
director_stats = (
    df[df['director'] != 'Unknown']
    .groupby('director')
    .agg(
        movie_count=('title', 'count'),
        avg_rating=('vote_average', 'mean')
    )
    .query('movie_count >= 3')
    .sort_values('avg_rating', ascending=False)
    .head(10)
    .round(2)
)
print(director_stats)

# Genre average rating
print("\n=== AVERAGE RATING BY GENRE ===")
print(
    df_genres[df_genres['vote_average'] > 0]
    .groupby('genre')['vote_average']
    .mean()
    .sort_values(ascending=False)
    .round(2)
)
```

```python
# Question 8 — Investment recommendation

# Safe bet profile — Horror and Thriller ROI vs flop rate
print("=== HORROR AND THRILLER PROFILE ===")
horror_thriller = financial_genres[
    financial_genres['genre'].isin(['Horror', 'Thriller'])
]
print(
    horror_thriller.groupby('genre').agg(
        avg_roi=('roi_pct', 'mean'),
        avg_budget=('budget', 'mean'),
        flop_rate=('was_flop', 'mean'),
        movie_count=('title', 'count')
    ).round(2)
)

# Best performing non-English languages by ROI
print("\n=== NON-ENGLISH LANGUAGES BY AVERAGE ROI ===")
print(
    non_english.groupby('language')['roi_pct']
    .mean()
    .sort_values(ascending=False)
    .head(8)
    .round(1)
)
```

## Phase 3 Complete — EDA and Business Questions

### Q1 — Netflix Library Overview
The dataset is remarkably evenly distributed — approximately 1,000 
movies per year from 2010 through 2024, with 2025 slightly lower at 
934 likely because the year is incomplete. This uniform distribution 
suggests the dataset was constructed with balanced sampling rather 
than reflecting actual Netflix addition patterns. English dominates 
the library at 9,534 movies, followed by French (1,054), Japanese 
(904), and Korean (876) — confirming Netflix's strong non-English 
content investment particularly in Asian markets.

### Q2 — Genre Volume vs Genre Profitability
There is a clear disconnect between what Netflix has most of and 
what actually makes money. Drama leads the library by volume (6,910 
appearances) but sits mid-table on ROI. Documentary and Music show 
astronomically high ROI figures — however these numbers are 
unreliable due to near-zero budgets recorded in the dataset making 
percentage returns mathematically extreme. Among genres with 
credible budget data, Horror (745% average ROI) and Thriller (488%) 
are the standout performers. The implication is that Netflix 
prioritizes Drama and Comedy for audience volume while Horror and 
Thriller deliver the strongest financial returns per dollar invested.

### Q3 — Budget vs Revenue
Budget and revenue show a strong positive correlation of 0.748 — 
higher spending generally produces higher revenue. However the ROI 
story tells the opposite: Low budget films (<$10M) show an average 
ROI of 196,549% driven by extreme outliers, while Mid, High, and 
Blockbuster tiers cluster between 149-214%. This means big budget 
films generate more absolute revenue but are not meaningfully more 
efficient investments than smaller productions. Spending more buys 
bigger gross — it does not buy proportionally better returns.

### Q4 — Flops vs Successes
36.7% of movies with financial data failed to recoup their budget 
— more than 1 in 3 films loses money. Flop risk drops dramatically 
as budget increases: Low budget films flop at 43.3%, Mid at 39.5%, 
High at 23.3%, and Blockbusters at only 10.4%. However audience 
ratings tell a nuanced story — flops average 6.29 vs successes at 
6.68, a difference of only 0.39 points. Box office failure and 
audience disappointment are not the same thing. A film can be 
genuinely well-received by audiences and still lose money, 
suggesting marketing, distribution, and timing are as important 
as quality.

### Q5 — Non-English Commercial Potential
Chinese (zh) leads non-English revenue by a significant margin at 
$224.9M average, followed by Kannada ($75.4M), Cantonese ($73.3M), 
Japanese ($66.7M), and Telugu ($63.9M). However volume context 
matters — French has the most movies with financial data (203) but 
lower average revenue ($not shown). Chinese and South Indian 
languages (Kannada, Telugu, Tamil) show high average revenue but 
from smaller sample sizes (65, 19, and 19 movies respectively), 
meaning results may be skewed by a few blockbusters. Korean (61 
movies, $41.9M average) represents the most balanced combination 
of volume and commercial performance among non-English markets.

### Q6 — Movie Quality Over Time
Audience ratings show a slow but consistent improvement from 6.15 
in 2011 to a peak of 6.50 in 2021, before settling around 6.41-6.46 
in recent years. The improvement is modest — less than half a point 
over 14 years — suggesting overall quality has been stable rather 
than dramatically improving. Popularity tells a more striking story: 
relatively flat through 2022, then a sharp spike to 74.33 in 2024 
before dropping to 35.59 in 2025. The 2024 spike likely reflects 
a small number of extremely popular titles distorting the average 
rather than a genuine industry-wide trend.

### Q7 — Directors and Genre Quality
Christopher Nolan is the only widely recognized director in the 
top 10 by average rating with 6 movies averaging 7.90 — the rest 
are specialists in animation, documentary, or niche genres with 
only 3 movies each, making their rankings statistically fragile. 
By genre, Documentary leads ratings at 7.10 followed by Music 
(6.96) and Animation (6.94). Horror sits at the bottom at 5.72 
confirming a consistent pattern — Horror generates strong financial 
returns but scores lowest on audience satisfaction. Quality and 
profitability point in opposite directions for Horror content.

### Q8 — Investment Recommendation
Based on the full analysis, the strongest risk-adjusted investment 
profile for a Netflix movie combines a mid-range budget between 
$10-50M, a Horror or Thriller genre, and consideration of Korean 
or Chinese language productions. Horror delivers 745% average ROI 
on an average budget of $14.7M — the lowest average spend among 
analyzed genres — while maintaining a flop rate of 35% which, 
though not low in absolute terms, is significantly better than the 
43% flop rate seen in unstructured low-budget productions. The 
strong budget-revenue correlation of 0.748 confirms that some 
minimum investment is necessary to generate meaningful returns, 
making the sub-$10M tier riskier despite its theoretical ROI 
figures. Korean content offers the best combination of sample 
volume and commercial performance among non-English markets at 
$41.9M average revenue across 61 verified titles. The data 
therefore points toward a mid-budget Korean or English-language 
Horror or Thriller as the most defensible investment strategy — 
high enough budget to ensure quality production, genre with proven 
ROI history, and language market with demonstrated commercial 
appetite.

The cleaned dataset contains 15,893 Netflix movies spanning 2010 
to 2025 across 19 genres and 20+ languages. All structural issues 
from the raw data have been resolved — wrong data types corrected, 
text standardized, and genuinely empty columns removed. 

Financial analysis is restricted to a verified subset of 3,540 
movies (22.3% of the library) where both budget and revenue records 
exist. The remaining 77.7% had zeros recorded for financial fields 
which have been converted to NaN. Any model trained on budget or 
revenue data should be built on financial_df only and findings 
should not be generalized to the full library.

Three important caveats for downstream modeling:
1. vote_count, vote_average, and rating contain genuine zeros 
   representing unrated movies — these should be filtered out 
   for any rating-based model rather than treated as nulls
2. The genres column contains multiple values per row — 
   exploding is required before using genre as a model feature
3. Documentary and Music ROI figures are extreme outliers driven 
   by near-zero recorded budgets and should be excluded from 
   any financial modeling or treated as a separate category

The dataset is ready for visualization and machine learning 
with the caveats above clearly understood.
